# 1. Lists

## Introducing Lists (Automotive Context)
A **list** is an ordered collection. In vehicle software, use lists for:
- Sensor readings (RPM, coolant temp) sampled over time
- **CAN IDs**, **DTCs** (Diagnostic Trouble Codes)
- **ECU** names and software versions
- Calibration values (offsets, thresholds)

### Example


```python

# Example: RPM readings and DTCs captured during a test drive
rpm_readings = [820, 1500, 2100, 1800, 2300]
dtc_list = ["P0300", "U0100", "P0420"]
can_ids = [0x100, 0x200, 0x201, 0x321]

rpm_readings, dtc_list, [hex(cid) for cid in can_ids]
```


## Naming and Defining a List
- Use **plural** names for collections (`readings`, `frames`, `dtcs`).
- Lists use square brackets `[]` and allow **mixed** types (keep it clean in production!).


```python

calibration = ["TPS_OFFSET", 0.02, "deg", True]  # name, value, unit, valid-flag
calibration
```


## Accessing One Item in a List
Indexing starts at **0**. Negative indexes access from the end.


```python

print("First CAN ID:", hex(can_ids[0]))
print("Last DTC:", dtc_list[-1])
```

# 2. Lists and Looping

## Accessing All Elements in a List
Use a `for` loop to iterate (e.g., process all frames in a buffer).


```python

for rpm in rpm_readings:
    print(f"RPM sample: {rpm}")
```


## Enumerating a List
Use `enumerate()` to get the **index and value** — ideal for cycle counts, step IDs, or test cases.


```python

for idx, dtc in enumerate(dtc_list):
    print(f"#{idx}: DTC={dtc}")
```

# 4. Removing Items from a List

## Removing items by position


```python

# Delete the second CAN ID by index
ids = [0x100, 0x200, 0x201, 0x321]
del ids[1]
[hex(x) for x in ids]
```


## Removing items by value


```python

# Remove a specific DTC by its code (first occurrence only)
dtc_list.remove("P0300")
dtc_list
```


## Popping items
`pop()` removes **and returns** an item: LIFO by default, or `pop(0)` for FIFO buffers.


```python

frames = [0x100, 0x200, 0x201]
last = frames.pop()     # LIFO
first = frames.pop(0)   # FIFO
hex(last), hex(first), [hex(x) for x in frames]
```


# 6. Slicing a List
Slicing `[start:end:step]` is perfect for extracting **drive cycle phases** (idle, accel, cruise).


```python

cycle_rpm = list(range(700, 3500, 200))  # synthetic profile
# first three samples, mid section, every second sample
cycle_rpm[:3], cycle_rpm[3:8], cycle_rpm[::2]
```


## Copying a list (by value vs reference)
Use `copy()` or `[:]` to avoid editing the original calibration set.


```python

original = [0x100, 0x200, 0x201]
copy_ok = original.copy()        # or original[:]
copy_bad = original              # reference!

copy_ok.append(0x321)
copy_bad.append(0x555)

[hex(x) for x in original], [hex(x) for x in copy_ok]
```

# 8. List Comprehensions

# 12. Sets
Sets hold **unique** items — great for finding unique CAN IDs or DTCs.


```python

logged = {0x100, 0x200, 0x200, 0x321}
supported = {0x100, 0x201, 0x321, 0x400}
unique_logged = set(logged)
missing = supported.difference(unique_logged)
unique_logged, [hex(x) for x in missing]
```


## Basic Operations on Sets


```python

a = {"ABS", "PCM", "TCU"}
b = {"TCU", "BCM", "IC"}
a.union(b), a.intersection(b), a.difference(b)
```
