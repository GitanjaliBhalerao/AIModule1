5. NumPy Example – Sales Revenue Calculation
Explanation

Element‑wise multiplication
Efficient numerical computation
Finding Maximum Revenue
Explanation

argmax() returns index of maximum value
Useful for ranking, optimization, analytics

```python
import numpy as np

#raw data
products=np.array(['Laptop','Smartphones','Tablets','Headphones','keyboard','Mouse','TV'])
units_sold=np.array([100,150,50,250,300,100,120])
prices=np.array([50000.00,20000.00,25000.00,2000.00,1000.00,500.00,60000.00])

#calculate revenue
revenues=units_sold*prices
print(revenues)
```

```python
#find highest revenue
max_index=np.argmax(revenues)
print(max_index)
print('Highest grossing product ',(revenues[max_index]))
#products[0]
```

```python
large_quantity=np.argmax(units_sold)
print('Large quantity : index ',large_quantity)
print('Large quantity : value ',(units_sold[large_quantity]))
```

Finding Row with Maximum Revenue
Explanation

idxmax() → index of highest value
loc[] → access rows by label (not position)

```python
# loc function is used to access rows and columns by their names not by their index numbers
# df.loc[row_label,column_label]
# idxmax = index(label) of the first occurence of the maximum value in series or dataframe
max_row=data.loc[data['Revenue'].idxmax()]
#print(data)
print(max_row['Product'])
print(max_row['units_sold'])
```

9. Transpose Data (Rows ↔ Columns)
Why transpose?

Required when data is stored vertically
Makes calculations easier

```python

import pandas as pd
import numpy as np

# Load CSV
data = pd.read_csv("sales_transpose.csv")

# Set 'product' as index and transpose
data_t = data.set_index('product').T
print("\nCorrect Transposed Data:")
print(data_t)

# Calculate revenue
data_t['revenue'] = data_t['units_sold'].astype(int) * data_t['Price'].astype(int)
print("\nData with Revenue:")
print(data_t)
```

```python
max_row=data_t.loc[data_t['revenue'].idxmax()]
print(data)
max_row
```

## Pandas Undo or rollback changes using  pandoras

10. Undo / Redo DataFrame Changes (Pandoras)
Why Pandoras?

Adds undo / redo capability
Helpful for experimentation & learning

```python
pip install pandoras
```

```python
import pandoras as pd  # Now PandorasDataFrame replaces pd.DataFrame

# Create a DataFrame
df = pd.DataFrame({"A": [1, 2, 3], "B": [4, None, 6]})

# Drop a column
df.drop(columns=["B"], inplace=True)
print("After drop:\n", df)

# Undo the drop
df.undo()
print("After undo:\n", df)

# Redo the drop
df.redo()
print("After redo:\n", df)
```

12. Subplots & Dashboard Export
Explanation

```python
#subplots 

fig,ax=plt.subplots(2,2)
ax[0,0].plot([1,2,3],[1,4,9])
ax[0,1].bar(['A','B'],[10,20])
ax[1,0].scatter([1,2,3],[3,2,1])
ax[1,1].hist([1,2,2,3,3,3])
plt.tight_layout()
#plt.savefig('report.png',dpi=300,bbox_inches='tight')
plt.savefig('dashboard.pdf',format='pdf',bbox_inches='tight')
plt.show()


```
