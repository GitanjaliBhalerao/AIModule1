```python
#try block with multiple excepts
try:
    n1=float(input("enter n1:"))
    n2=float(input("enter n2:"))
    result=n1/n2
    print(result)
except ZeroDivisionError:
    print('Division by zero is not possible. try executing the same code with non-zero values')
except ValueError as err:
    print("Expecting numeric input ,",err)
else:
    print("Code executely successfully, executing just after try block")
finally:
    print("executes always! Its used for closure tasks")
```

```python
#custom Exception ==> user defined exceptions
import random

class Error(Exception):
    pass

class TooSmallError(Error):
    pass

class TooLargeError(Error):
    pass

num=random.randint(1,100) #number between 1 to 100
print(num)

while True:
    try:
        user_num=int(input("number:"))

        if user_num<num:
            raise TooSmallError('Value is too small,try again...')
        elif user_num>num:
            raise TooLargeError('Value is too large, try again...')
        else:
            break

    except TooSmallError as err:
            print(err)
    except TooLargeError as err:
            print(err)


print("You have guess it right")
    



```
